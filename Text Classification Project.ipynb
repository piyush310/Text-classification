## Loading Stopwords:

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords = set(stopwords.words('spanish'))
```

```python
stopwords
```

```python
directory=[i for i in os.listdir('./20_newsgroups/')]
```

```python
directory
```

## Making Dictionary of words and there Frequency(ex.{'cool':256,'root':4569}):

```python
result={}
for i in range(len(directory)):
    files=os.listdir('./20_newsgroups/' + directory[i])
    for j in range(len(files)):
        path='./20_newsgroups/'+directory[i]+'/'+files[j]
        text=open(path,'r',errors='ignore').read()
        for k in text.split():
            if(len(k)!=1):
                if( not k.lower() in stopwords):
                    if(result.get(k.lower()) != None):
                        result[k.lower()]+=1
                    else:
                        result[k.lower()]=1
```

## Sorting the dictionary and taking top 2000 words with highest frequency


```python
import operator
sorted_result = sorted(result.items(), key=operator.itemgetter(1),reverse=True)
```

```python
result1={}
a=sorted_result[2000][1]
for i in sorted_result:
    result1[i[0]]=i[1]
    if(i[1]<a):
        break
```

```python
feature=list(result1.keys())
```

## Process of Loading the data into a dataFrame:

```python
df=pd.DataFrame(columns=feature)
```

```python
for i in range(len(directory)):
    files=os.listdir('./20_newsgroups/'+directory[i])
    for j in range(len(files)):
        path='./20_newsgroups/'+directory[i]+'/'+files[j]
        text=open(path,'r',errors='ignore').read()
        df.loc[len(df)] = np.zeros(len(feature))
        for k in text.split():
            if(k.lower() in feature):
                df[k.lower()][len(df)-1]+=1
```

## Multinomial NavieBayes From Scratch:

```python
f=list(df)
```

#### Making dictionary of words and there Frequency:

```python
def fit(x_train,y_train):
    result={}
    class_count=set(y_train)
    for current_class in class_count:
        t=(y_train == current_class)
        x_curr=x_train[t]
        y_curr=y_train[t]
        result["total_data"]=len(y_train)
        sum=0
        result[current_class]={}
        for i in range(len(f)):
            result[current_class][f[i]]=x_curr[:,i].sum()
            sum=sum+result[current_class][f[i]]
        result[current_class]["total_count"]=sum
    return result
```

#### Finding the Probability:

```python
def probability(dictionary, x, current_class):
    class_prob = np.log(dictionary[current_class]["total_count"]) - np.log(dictionary["total_data"])
    total_prob = class_prob
    for i in range(len(x)):
        word_count = dictionary[current_class][f[i]] + 1     
        total_count = dictionary[current_class]["total_count"] + len(f)
        word_prob = np.log(word_count) - np.log(total_count)
        for j in range(int(x[i])):
            total_prob += word_prob
    return total_prob
```

#### Predicting the which class data belong:

```python
def predictSinglePoint(dictionary,x):
    classes=dictionary.keys()
    best_prob=-1000
    best_class=-1
    first_run=True
    for current_class in classes:
        if(current_class == "total_data"):
            continue
        prob=probability(dictionary,x,current_class)
        if(first_run or prob>best_prob):
            best_prob=prob
            best_class=current_class
        first_run=False
    return best_class
```

```python
def predict(dictionary,x_test):
    y_predict1=[]
    for i in range(len(x_test)):
        pred=predictSinglePoint(dictionary,x_test[i,:])
        y_predict1.append(pred)
    return y_predict1
```

```python
dictionary=fit(x_train,y_train)
```

#### Result of prediction from scratch

```python
y_predict1=predict(dictionary,x_test)
```
